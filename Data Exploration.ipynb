# Feature 3 - Texture

```python
texture = train.iloc[:,130:]
```

```python
plt.bar(range(64),texture.mean())
plt.xlabel('Texture 1-64')
plt.ylabel('Texture')
plt.title('Mean for Each of the Texture Attribute', fontsize=20)
plt.xlim(0,64)
```

```python
plt.bar(range(64),990-(texture.astype(bool).sum(axis=0)))
plt.xlabel('Texture 1-64')
plt.ylabel('Count of Zeros')
plt.title('Number of Zeros per Texture Attribute', fontsize=20)
plt.xlim(0,64)
```

```python
print ' The average number of zeros in the texture attribute for every leaf sample is:', (64-(texture.astype(bool).sum(axis=1))).mean()
```

```python
print 'The maximum value for the texture feature is ', texture.max().max(), 'the minimum value for the texture feature is ', texture.min().min()
```
